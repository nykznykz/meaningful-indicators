```python
def calc_propn_ratio(caption):
    propn = 0
    punct = 0
    for token in caption:
#         print('token: %16s | token.pos: %8s | token.pos_: %s' % (token, token.pos, token.pos_))
        if token.pos_ == "PROPN":
            propn += 1
        if token.pos_ == "PUNCT":
            punct += 1
    return propn /(len(caption) - punct)
```

```python
calc_propn_ratio(test)
```

```python
for caption in captions:
    caption = nlp(caption)
    if len(caption.ents) > 3:
        print(caption)
        print("----------------------")
        
```

```python
for caption in captions:
    caption = nlp(caption)
    if calc_propn_ratio(caption) > 0.02:
        print("%s | %s \n" % (caption, calc_propn_ratio(caption)))
```

## Analysis of Hashtags

```python
tagged_ig_users = []
for caption in captions:
    if "@" in caption:
        caption = nlp(caption)
        for token in caption:
            if token.text[0] == "@" and len(token.text) > 1:
                tagged_ig_users.append(token.text)
```

```python
from collections import Counter

counts = Counter(tagged_ig_users)
```

```python
counts.most_common()
```

```python
n_sponsored = 0
for caption in captions:
    if "$" in caption:
        n_sponsored += 1
        caption = nlp(caption)
        print(caption)
        print("----------------------")
        
```
